Построим распределения и проведём статистические тесты


```python
from tabulate import tabulate
from scipy import stats
```

```python
def chi2_contingency_test(data, col, target, alpha=0.01, to_print=False):
    if to_print:
        print(f'Chi2 contingency test for |{col}|')

    table = pd.crosstab(data[col], data[target])

    if to_print:
        if data[col].nunique() < 15:
            print(tabulate(table, headers='keys', tablefmt='psql'))

    table = table.to_numpy()
    res = stats.chi2_contingency(table)

    if to_print:
        print(f'ndf = {res.dof}')
        print(f'statistics = {res.statistic}')
        print(f'p-value = {res.pvalue}')
        if res.pvalue < alpha:
            print('H0 rejected')
        else:
            print('H0 accepted')

    return res.pvalue

def cat_feature_test(data: pd.DataFrame, col: str, target: str) -> None:
    plt.figure(figsize=(8, 5))
    axis = plt.subplot(1, 1, 1)
    sns.countplot(data=data,
                  x=col,
                  hue=target)
    title = f"chi2 test p-value = {round(chi2_contingency_test(data, col, target), 3)}"
    axis.set(ylabel='Count', title=title)
    axis.tick_params('x', length=0, width=0, which='both')
    if data[col].nunique() > 3:
        axis.tick_params('x', length=0, width=0, which='both', rotation=45)
```

Один из способов закодировать категориальные признаки с опцией "No internet service"

```python
def transform_features(df: pd.DataFrame, cols_internet: list) -> pd.DataFrame:
  df_new = df.copy()

  df_new['TotalSpent'] = np.log(df_new['TotalSpent'])
  df_new['Sex'] = df_new['Sex'].replace({'Male': 0, 'Female': 1})
  df_new['HasPartner'] = df_new['HasPartner'].replace({'No': 0, 'Yes': 1})
  df_new['HasChild'] = df_new['HasChild'].replace({'No': 0, 'Yes': 1})
  df_new['HasPhoneService'] = df_new['HasPhoneService'].replace({'No': 0, 'Yes': 1})
  df_new['IsBillingPaperless'] = df_new['IsBillingPaperless'].replace({'No': 0, 'Yes': 1})

  df_new['HasInternet'] = (~(df_new["HasInternetService"] == 'No')).astype(int)
  df_new['HasInternetService'] = np.where(df_new["HasInternetService"] != 'No', df_new["HasInternetService"], 'DSL')
  df_new['HasInternetService'] = df_new['HasInternetService'].replace({'DSL': 0, 'Fiber optic': 1})

  for i in cols_internet:
    df_new[i] = df_new[i].replace({'No': 0, 'No internet service': 1, 'Yes': 1}) # 'No', 'Yes', 'No internet service'

  return df_new
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV, cross_validate
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix, auc
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline, Pipeline
```

```python
def plot_roc_auc(y_test: np.array, y_pred_proba: np.array) -> None:
    """
    Построение ROC-кривой и рассчёт площади под ней.

    Parameters
    ----------
    y_test : np.array
        Вектор истинных значения меток (0 или 1)
    y_pred_proba : np.array
        Вектор псевдоверояностей значения 1

    Returns
    -------
    None
    """
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc = auc(fpr, tpr)

    # Plot ROC Curve
    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (area = %0.3f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='lower right')
    plt.show()
```

```python
X = df_train_new.drop([target_col], axis=1)
y = df_train_new[target_col]
```

```python
cat_cols_new = []
num_cols_new = []

for i, t in zip(df_train_new.dtypes.index, df_train_new.dtypes):
  if i != target_col:
    if t == object:
      cat_cols_new.append(i)
    else:
      num_cols_new.append(i)
```

Собираем пайплайн для модели с MinMaxScaler для числовых признаков и OneHotEncoder для категориальных. В качетве классификатора логистическая регрессия с L2 регуляризацией

```python
scaler = MinMaxScaler()
onehot_encoder = OneHotEncoder()
scaler = MinMaxScaler()
logreg = LogisticRegression()

# отдельно обработка категориальных и числовых признаков
preprocessor = ColumnTransformer(
    transformers=[('ohe', onehot_encoder, cat_cols_new),
                  ('scaler', scaler, num_cols_new)
                 ])

# pipeline
pipeline_logreg = Pipeline(steps=[('preprocessor', preprocessor),
                                  ('scaler', scaler),
                                  ('clf', logreg)
                                 ])
```

Посмотрим на метрики модели с гиперепараметрами по умолчанию на кросс-валидации

```python
scoring = ['f1', 'roc_auc']
cv_results = cross_validate(pipeline_logreg, X, y, cv=5, scoring=scoring)
print(f"mean roc auc score = {cv_results['test_roc_auc'].mean()}")
print(f"mean f1 score = {cv_results['test_f1'].mean()}")
```

Нормально, попробуем улучшить с помощью подбора гиперпараметров на сетке. В частности можно подбирать коэффициент регуляризации и тип регуляризации

```python
params = {
    'clf__solver': ['lbfgs', 'liblinear', 'sag', 'newton-cholesky'],
    'clf__penalty': ['l2'],
    'clf__C': [0.1, 1, 5, 10, 20, 50, 100, 150, 200]
    }

search = GridSearchCV(pipeline_logreg, params, cv=5, scoring='roc_auc', n_jobs=-1)
search.fit(X, y)
```

```python
from catboost import CatBoostClassifier, Pool
import optuna

optuna.logging.set_verbosity(optuna.logging.WARN)

def logging_callback(study, frozen_trial):
    previous_best_value = study.user_attrs.get("previous_best_value", None)
    if previous_best_value != study.best_value:
        study.set_user_attr("previous_best_value", study.best_value)
        print(
            "Trial {} finished with best value: {} and parameters: {}. ".format(
            frozen_trial.number,
            frozen_trial.value,
            frozen_trial.params,
            )
        )
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=44, test_size=0.2)
print(f'Train shape: {X_train.shape}')
print(f'Validation shape: {X_val.shape}')
```

```python
cat_features_index = []

for i, col in enumerate(X_train.columns):
  if X_train[col].dtype == object:
    cat_features_index.append(i)

print(cat_features_index)
```

```python
train_pool = Pool(X_train, y_train,
                  cat_features=cat_features_index)
val_pool = Pool(X_val, y_val,
                cat_features=cat_features_index)


model_cb = CatBoostClassifier()

model_cb.fit(X_train, y_train,
            cat_features=cat_features_index,
            logging_level='Silent',
            eval_set=val_pool,
            early_stopping_rounds=20)
```

```python
y_pred_proba = model_cb.predict_proba(X_val)[:, 1]
plot_roc_auc(y_val, y_pred_proba)
```

Возможно, я перемудрили с кодированием колонок в начале ноутбука, когда придумавал как бы уменьшить число категориальных признаков, поэтому catboost теперь перформит так себе. Но я всё ещё склоняюсь к тому, что мои дейтвия были правильными. Так или иначе, возможно, тюннинг гипермаратров катбуста, позволит добиться скора хотя бы выше, чем у линейной модели

```python
def objective(trial):

    param = {
        "eval_metric": 'AUC',
        "verbose": False,
        "n_estimators": trial.suggest_int("n_estimators", 10, 300),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 1.0, step=0.01),
        "subsample": trial.suggest_float("subsample", 0.05, 1.0, step=0.01),
        "colsample_bylevel": trial.suggest_float("colsample_bylevel", 0.05, 1.0, step=0.01),
        "min_data_in_leaf": trial.suggest_int("min_data_in_leaf", 1, 100),
        "l2_leaf_reg": trial.suggest_loguniform("l2_leaf_reg", 1, 100),
        "bagging_temperature": trial.suggest_loguniform("bagging_temperature", 0.1, 20.0)
        #"boosting_type": trial.suggest_categorical('boosting_type', ['Ordered', 'Plain'])
    }

    estimator = CatBoostClassifier(**param)

    estimator.fit(X_train, y_train,
                  cat_features=cat_features_index,
                  logging_level='Silent',
                  )
    pred_proba = estimator.predict_proba(X_val)[:, 1]

    return roc_auc_score(y_val, pred_proba)


study = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler())
study.optimize(objective, n_trials=1000, show_progress_bar=True, n_jobs=-1)
```

```python
for k, v in study.best_params.items():
  print(k, v, sep=': ')
```

```python
model_cb = CatBoostClassifier(**study.best_params, verbose=False)
model_cb.fit(X_train, y_train,
            cat_features=cat_features_index,
            logging_level='Silent')

y_pred_proba = model_cb.predict_proba(X_val)[:, 1]
plot_roc_auc(y_val, y_pred_proba)
```
